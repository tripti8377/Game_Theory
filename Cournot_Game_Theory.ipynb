```python
import matplotlib.pyplot as plt
import numpy as np

def plot_two_lines(a, c1, c2):
    """
    Plots two lines given parameters and calculates their intersection point.

    Parameters:
    a (float): An arbitrary parameter affecting the slope or intercept.
    c1 (float): Parameter affecting the first line.
    c2 (float): Parameter affecting the second line.
    """
    # Generate x values
    x = np.linspace(0, 100, 10000)  # 400 points between 0 and 100

    # Compute y values for each line
    y1 = (-1/2) * x + (1/2) * a + (-1/2) * c1
    y2 = (x + (1/2) * c2 - (1/2) * a) * (-2)

    # Calculate the intersection point
    # Solve (-1/2)x + (1/2)a - (1/2)c1 = (x + (1/2)c2 - (1/2)a)*(-2)
    # Rearranging, we get x_intersect = (c1 - c2 + a) / (3/2)
    x_intersect = (c1 - 2*c2 + a) / 3
    y_intersect = (c2 - 2*c1 + a) / 3

    # Create a new figure
    plt.figure(figsize=(8, 6))

    # Plot the first line
    plt.plot(x, y1, color='r', label='Line 1')

    # Plot the second line
    plt.plot(x, y2, color='b', label='Line 2')

    # Plot the intersection point
    plt.scatter(x_intersect, y_intersect, color='green', zorder=5)
    plt.annotate(f'({x_intersect:.2f}, {y_intersect:.2f})',
                 xy=(x_intersect, y_intersect),
                 xytext=(x_intersect , y_intersect ),
                 textcoords='offset points',
                 arrowprops=dict(facecolor='black', shrink=0.05),
                 fontsize=12)

    # Add title and labels
    plt.title('Plot of Two Lines with Intersection Point')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # Add a legend
    plt.legend()

    # Add grid
    plt.grid(True)

    # Display the plot
    plt.show()

# Example usage
c1 = float(input("Enter the number of cupcakes made by Jim c1: "))
h1 = float(input("Enter the amount of bread and icing used by Jim h1: "))

c2 = float(input("Enter the number of cupcakes made by Jessie c2: "))
h2 = float(input("Enter the amount of bread and icing used by Jessie h2: "))

a = float(input("Enter parameter a: "))

c = c1 + c2
p = a-c
if(c1 * p - c1 * h1>=0):
  print("Payment earned by Jim = ", c1 * p - c1 * h1)
else:
  print("Payment earned by Jim = 0")
if(c2 * p - c2 * h2>=0):
  print("Payment earned by Jessie = ", c2 * p - c2 * h2)
else:
  print("Payment earned by Jessie = 0")

br_jim= (a-c2-h1)/2
br_jessie= (a-c1-h2)/2
print("Best response i.e given Jessie makes c2 cupcakes how many cupackes should Jim make should make to get higher pay : " , br_jim)
print ("Payment earned if above number of cupcakes are made by Jim : ", (br_jim*p)-br_jim*h1)
print("Best response i.e given Jim makes c1 cupcakes how many cupackes should Jessie make should make to get higher pay : ", br_jessie)
print("Payment earned if above number of cupcakes are made by Jessie :  " , (br_jessie*p)-br_jessie*h2)
m_br_jim= (a+h1-2*h2)/3
m_br_jessie= (a+h2-2*h1)/3
print("Mutual best response (number of cupcakes to be made) by Jim : " , m_br_jim)
print("Payment earned by Jim if mutually best response is achieved : " , (m_br_jim*p)-m_br_jim*h1)
print("Mutual best response (number of cupcakes to be made) by Jessie : " , m_br_jessie)
print("Payment earned by Jessie if mutually best response is achieved : ", (m_br_jessie*p)-m_br_jessie*h2)

plot_two_lines(a, h1, h2)
```
